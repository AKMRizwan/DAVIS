```python
# ecommerce_data_analysis.py

import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Database Connectivity
def connect_to_database():
    connection = mysql.connector.connect(
        user='root',
        password='root',
        host='localhost',
        database='ecommerce'
    )
    return connection

# Fetch data from the database
def fetch_data():
    connection = connect_to_database()
    cursor = connection.cursor()

    # Query data from the 'customer' table
    cursor.execute('SELECT * FROM customer')
    customer_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    # Query data from the 'product' table
    cursor.execute('SELECT * FROM product')
    product_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    # Query data from the 'order_details' table
    cursor.execute('SELECT * FROM order_details')
    order_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    cursor.close()
    connection.close()

    return customer_data, product_data, order_data

# Data Cleaning
def clean_data(customer_data, product_data, order_data):
    # Check for missing values
    print("Customer Data Missing Values:\n", customer_data.isnull().sum())
    print("Product Data Missing Values:\n", product_data.isnull().sum())
    print("Order Data Missing Values:\n", order_data.isnull().sum())

    # Handle missing values if any (example: filling with a placeholder)
    customer_data.fillna('Unknown', inplace=True)
    product_data.fillna({'stock': 0}, inplace=True)
    order_data.fillna({'quantity': 0}, inplace=True)

    # Convert columns to appropriate data types
    customer_data['pin_code'] = customer_data['pin_code'].astype(str)
    order_data['order_date'] = pd.to_datetime(order_data['order_date'], errors='coerce')

    return customer_data, product_data, order_data

# Exploratory Data Analysis (EDA)
def perform_eda(customer_data, product_data, order_data):
    sns.set(style="whitegrid")

    # Customer Analysis
    print("\nCustomer Analysis:\n")

    # Total number of customers city-wise
    city_counts = customer_data['city'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=city_counts.index, y=city_counts.values, hue=city_counts.index, palette="viridis", legend=False)
    plt.title("Total Number of Customers City-Wise")
    plt.xlabel("City")
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Total Number of Customers City-wise:")
    print(city_counts)

    # Most frequent customers based on order history
    customer_order_counts = order_data['customer_id'].value_counts()
    frequent_customers = customer_data[customer_data['customer_id'].isin(customer_order_counts.head(10).index)]
    print("Most Frequent Customers:\n", frequent_customers)

    # Product Analysis
    print("\nProduct Analysis:\n")

    # Total number of products by category
    category_counts = product_data['category'].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=category_counts.index, y=category_counts.values, hue=category_counts.index, palette="plasma", legend=False)
    plt.title("Total Number of Products by Category")
    plt.xlabel("Category")
    plt.ylabel("Number of Products")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Total Number of Products Available by Category:")
    print(category_counts)

    # Distribution of products across sub-categories
    sub_category_counts = product_data['sub_category'].value_counts().head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=sub_category_counts.index, y=sub_category_counts.values, hue=sub_category_counts.index, legend=False)
    plt.title('Distribution of Products Across Sub-categories')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Distribution of Products Across Sub-categories:")
    print(sub_category_counts)

    # Products with low stock levels
    low_stock = product_data[product_data['stock'] < 10].head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=low_stock['product_name'], y=low_stock['stock'], hue=low_stock['product_name'], legend=False)
    plt.title('Products with Low Stock Levels')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Products with Low Stock Levels:")
    print(low_stock)

    # Average, maximum, and minimum prices for products
    plt.figure(figsize=(8, 6))
    sns.histplot(product_data['selling_price'], kde=True, color="blue")
    plt.title("Distribution of Product Selling Prices")
    plt.xlabel("Selling Price")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

    price_stats = product_data[['selling_price']].agg(['mean', 'max', 'min'])
    print("Average, Maximum, and Minimum Prices for Products:")
    print(price_stats)

    # Order Analysis
    print("\nOrder Analysis:\n")

    # Top 10 orders product-wise
    top_orders = order_data.groupby('product_id')['quantity'].sum().nlargest(10)
    top_order_products = product_data[product_data['product_id'].isin(top_orders.index)]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_order_products['product_name'], y=top_orders.values, hue=top_order_products['product_name'], legend=False)
    plt.title('Top 10 Orders Product Wise')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Top 10 Orders Product Wise:")
    print(top_orders)

    # Order status distribution
    order_status = order_data['order_status'].value_counts()
    plt.figure(figsize=(10, 5))
    sns.barplot(x=order_status.index, y=order_status.values, hue=order_status.index, legend=False)
    plt.title('Order Status Distribution')
    plt.tight_layout()
    plt.show()

    print("Order Status Distribution:")
    print(order_status)

    # Most popular products based on order quantity
    popular_products = order_data.groupby('product_id')['quantity'].sum().nlargest(10)
    popular_product_names = product_data[product_data['product_id'].isin(popular_products.index)]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=popular_product_names['product_name'], y=popular_products.values, hue=popular_product_names['product_name'], legend=False)
    plt.title('Most Popular Products Based on Order Quantity')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Most Popular Products Based on Order Quantity:")
    print(popular_products)

    # Sales Analysis
    print("\nSales Analysis:\n")

    # Total revenue generated from orders product-wise
    revenue_product = order_data.groupby('product_id')['total_price'].sum().nlargest(5)
    revenue_product_data = product_data[product_data['product_id'].isin(revenue_product.index)]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=revenue_product_data['product_name'], y=revenue_product.values, hue=revenue_product_data['product_name'], legend=False)
    plt.title('Total Revenue Generated from Orders Product Wise')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Total Revenue Generated from Orders Product Wise:")
    print(revenue_product)

    # Total revenue generated from all orders
    total_revenue = order_data['total_price'].sum()
    print(f"Total Revenue Generated from All Orders: ${total_revenue:.2f}")

    # Revenue category-wise percentage
    category_revenue = order_data.merge(product_data, on='product_id').groupby('category')['total_price'].sum()
    category_percentage = (category_revenue / total_revenue) * 100
    plt.figure(figsize=(10, 5))
    sns.barplot(x=category_percentage.index, y=category_percentage.values, hue=category_percentage.index, legend=False)
    plt.title('Revenue Category Wise Percentage')
    plt.xticks(rotation=45)
    plt.ylabel('Percentage')
    plt.tight_layout()
    plt.show()

    print("Total Revenue Product Category Wise Percentage:")
    print(category_percentage)

    # Most profitable products based on original and selling prices
    product_data['profit'] = product_data['selling_price'] - product_data['original_price']
    profitable_products = product_data.nlargest(10, 'profit')
    plt.figure(figsize=(10, 5))
    sns.barplot(x=profitable_products['product_name'], y=profitable_products['profit'], hue=profitable_products['product_name'], legend=False)
    plt.title('Most Profitable Products')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Most Profitable Products Based on Difference Between Original and Selling Prices:")
    print(profitable_products.head(10))

    # Customer Order Patterns
    print("\nCustomer Order Patterns:\n")

    # Product names with highest and lowest order quantities
    order_quantities = order_data.groupby('product_id')['quantity'].sum()
    highest_order_product = product_data[product_data['product_id'].isin([order_quantities.idxmax()])]
    lowest_order_product = product_data[product_data['product_id'].isin([order_quantities.idxmin()])]
    print("Product with Highest Order Quantity:", highest_order_product['product_name'].values[0])
    print("Product with Lowest Order Quantity:", lowest_order_product['product_name'].values[0])

    # Customers with highest and lowest order quantities
    customer_order_totals = order_data.groupby('customer_id')['quantity'].sum()
    top_customer = customer_data[customer_data['customer_id'] == customer_order_totals.idxmax()]
    bottom_customer = customer_data[customer_data['customer_id'] == customer_order_totals.idxmin()]
    print("Customer with Highest Order Quantity:", top_customer['name'].values[0])
    print("Customer with Lowest Order Quantity:", bottom_customer['name'].values[0])

    # Most preferred payment modes
    payment_modes = order_data['payment_mode'].value_counts()
    plt.figure(figsize=(10, 5))
    sns.barplot(x=payment_modes.index, y=payment_modes.values, hue=payment_modes.index, legend=False)
    plt.title('Most Preferred Payment Modes')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Most Preferred Payment Modes:")
    print(payment_modes)

    # Time-Based Analysis
    print("\nTime-Based Analysis:\n")

    # Month-wise total sales
    order_data['month'] = order_data['order_date'].dt.to_period('M')
    month_sales = order_data.groupby('month')['total_price'].sum().reset_index()
    month_sales['month'] = month_sales['month'].astype(str)
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=month_sales, x='month', y='total_price', marker='o')
    plt.title('Month Wise Total Sales')
    plt.xticks(rotation=45)
    plt.ylabel('Total Sales')
    plt.xlabel('Month')
    plt.tight_layout()
    plt.show()

    print("Month Wise Total Sales:")
    print(month_sales)

    # Month and year wise total sales
    year_month_sales = order_data.groupby(order_data['order_date'].dt.to_period('M'))['total_price'].sum().reset_index()
    year_month_sales['year_month'] = year_month_sales['order_date'].astype(str)
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=year_month_sales, x='year_month', y='total_price', marker='o')
    plt.title('Month and Year Wise Total Sales')
    plt.xticks(rotation=45)
    plt.ylabel('Total Sales')
    plt.xlabel('Month-Year')
    plt.tight_layout()
    plt.show()

    print("Month and Year Wise Total Sales:")
    print(year_month_sales)

    # Identify peak order date
    peak_order_date = order_data['order_date'].value_counts().idxmax()
    print(f"Peak Order Date: {peak_order_date.date()}")

    # Geographical Analysis
    print("\nGeographical Analysis:\n")

    # Distribution of customers across different cities
    city_distribution = customer_data['city'].value_counts().head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=city_distribution.index, y=city_distribution.values, hue=city_distribution.index, legend=False)
    plt.title('Distribution of Customers Across Different Cities')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Distribution of Customers Across Different Cities:")
    print(city_distribution)

    # Popular products in specific cities
    city_product_popularity = order_data.merge(customer_data, on='customer_id').merge(product_data, on='product_id')
    city_category_popularity = city_product_popularity.groupby(['city', 'category']).size().unstack().fillna(0).head(10)
    plt.figure(figsize=(15, 8))
    sns.heatmap(city_category_popularity, annot=True, cmap='YlGnBu', fmt='g')
    plt.title('Product Popularity Across Cities')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Products or Categories More Popular in Specific Cities:")
    print(city_category_popularity)

    # Product Performance
    print("\nProduct Performance:\n")

    # Top 10 best-selling products
    best_selling_products = order_data.groupby('product_id')['quantity'].sum().nlargest(10)
    best_selling_product_names = product_data[product_data['product_id'].isin(best_selling_products.index)]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=best_selling_product_names['product_name'], y=best_selling_products.values, hue=best_selling_product_names['product_name'], legend=False)
    plt.title('Top 10 Best-Selling Products')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Top 10 Best-Selling Products:")
    print(best_selling_products)

    # Top 10 slow-moving products based on low sales
    slow_moving_products = order_data.groupby('product_id')['quantity'].sum().nsmallest(10)
    slow_moving_product_names = product_data[product_data['product_id'].isin(slow_moving_products.index)]
    plt.figure(figsize=(10, 5))
    sns.barplot(x=slow_moving_product_names['product_name'], y=slow_moving_products.values, hue=slow_moving_product_names['product_name'], legend=False)
    plt.title('Top 10 Slow-Moving Products')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Top 10 Slow-Moving Products Based on Low Sales:")
    print(slow_moving_products)

    # Payment Analysis
    print("\nPayment Analysis:\n")

    # Display successful and pending payments order counts
    payment_status_counts = order_data['order_status'].value_counts()
    plt.figure(figsize=(10, 5))
    sns.barplot(x=payment_status_counts.index, y=payment_status_counts.values, hue=payment_status_counts.index, legend=False)
    plt.title('Payment Status Counts')
    plt.tight_layout()
    plt.show()

    print("Successful and Pending Payments Order Counts:")
    print(payment_status_counts)

# Main function to run the analysis
def main():
    customer_data, product_data, order_data = fetch_data()
    customer_data, product_data, order_data = clean_data(customer_data, product_data, order_data)
    perform_eda(customer_data, product_data, order_data)

if __name__ == "__main__":
    main()
```
